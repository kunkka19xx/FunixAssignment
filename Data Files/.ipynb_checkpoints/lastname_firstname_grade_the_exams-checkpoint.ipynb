```python
import numpy as np 
import pandas as pd
def openfile():
    while True:
        try:
            filename = str(input('Enter a class file to grade (i.e. class1 for class1.txt):'))
            filename = filename+'.txt'
            f = open(filename, "r")
            print('Successfully opened class1.txt')
            count=0
            list = []
            for row in f:
                row=row.strip().split(',')
                list.append(row)
                count = count+1
            print(count)
            valid_rows=0
            for i in range(len(list)):
                #start with N
                if(list[i][0][0]!='N'):
                    print("First character of row", i+1, "is invalid!")
                    print(list[i])
                # after N is 8 characters
                elif(len(list[i][0])!=9):
                    print("Id of row", i+1, "is invalid!")
                    print(list[i])
                # One row has 26 values
                elif(len(list[i]) < 26):
                    print("Answers number is",i+1,"too little!",len(list[i])-1)
                    print(list[i])
                elif (len(list[i]) > 26):
                    print("Aswers number is", i + 1, "too much!",len(list[i])-1)
                    print(list[i])
                else:
                    valid_rows=valid_rows+1
            print('File has',valid_rows,'valid rows')
            break
        except FileNotFoundError as fnf:
            print('File cannot be found.')
            print(fnf)
openfile()


# def valid_row(list):
#     for i in list:
```
